Mô hình Linear Regression cố gắng tạo ra một mối quan hệ tuyến tính giữa biến phụ thuộc (ví dụ: lương) và biến độc lập (ví dụ: kinh nghiệm).
Mục tiêu là tạo ra phương trình đường thẳng (y = mx + b) với sai số tối thiểu giữa giá trị thực tế và giá trị dự đoán.
Phương trình tổng quát cho Linear Regression đơn biến là: y = mx + b, và đa biến là: y = m1x1 + m2x2 + ... + mnxn + b.

Thì trong bài hôm này thì ta sẽ xây dựng một baseline model đơn giản và dễ thực hiện với biểu thức loss function như dưới .

$$
\mathcal{L}(\mathbf{w}) = \frac{1}{N} \sum_{i=1}^{N} (y_i - \mathbf{w}^{\intercal}\mathbf{x}_i) = \frac{1}{N}||\bar{\mathbf{X}}\mathbf{w} - \mathbf{y}||_{2}^{2}
$$

## 1. Imputation
Giờ Ta sẽ kiểm tra missing value và kiểm tra dữ liệu nhiều quá 50% thì sẽ drop dữ liệu đó, các dữ liệu thiếu và drop các biến không cần thiết.

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer

# Identify columns with more than 50% missing values
thresh = 0.5 * merged_df.shape[0]
drop_cols = merged_df.columns[merged_df.isnull().sum() > thresh]

# Drop columns with more than 50% missing values
merged_df = merged_df.drop(drop_cols, axis=1)

# Select only numeric columns for imputation
numeric_cols = merged_df.select_dtypes(include=np.number).columns
merged_df_numeric = merged_df[numeric_cols]

# Impute missing values with the mean for numeric columns only
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
merged_df_numeric = pd.DataFrame(imputer.fit_transform(merged_df_numeric), columns=merged_df_numeric.columns)

# Check for missing values in the numeric DataFrame
missing_values = merged_df_numeric.isnull().sum().sum()
print(f"Number of missing values after imputation: {missing_values}")


merged_df_numeric = merged_df_numeric.drop(['article_id','phone','price_per_m2'], axis=1)
```

# 2. Triển Khai Thực Hiện Model

##2.1 Implement code data science bằng thuật toán

Hàm fit:
$$
\begin{bmatrix} \hat{\beta_0} \\ \hat{\beta_1} \\ \vdots \\ \hat{\beta_p} \end{bmatrix} = (\mathbf{X}_\text{bias}^\top \mathbf{X}_\text{bias})^{-1} \mathbf{X}_\text{bias}^\top \mathbf{y}
$$
Hàm predict:
$$
\hat{y} = X_\text{bias} \cdot \begin{bmatrix} \hat{\beta_0} \\ \hat{\beta_1} \\ \vdots \\ \hat{\beta_p} \end{bmatrix}
$$
Hàm MSE
$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2
$$
hàm Rsquare
$$
R^2 = 1 - \frac{\sum_{i=1}^n (y_i - \hat{y}_i)^2}{\sum_{i=1}^n (y_i - \bar{y})^2}
$$

Hàm Predict

```python
import numpy as np
from sklearn.model_selection import cross_val_score, KFold
from sklearn.base import BaseEstimator, RegressorMixin

class LinearRegressionImp(BaseEstimator, RegressorMixin):
    def __init__(self, intercept_=None, coef_=None):
        self.coef_ = coef_
        self.intercept_ = intercept_
        self.is_fitted = False

    def fit(self, X, y):

        X_bias = np.hstack((np.ones((X.shape[0], 1)), X))

        theta_best = np.linalg.pinv(X_bias.T.dot(X_bias)).dot(X_bias.T).dot(y)

        self.intercept_ = theta_best[0]
        self.coef_ = theta_best[1:]
        self.is_fitted = True

    def predict(self, X):
        if not self.is_fitted:
            raise ValueError("Model is not fitted yet. Please call 'fit' with appropriate arguments before using this estimator.")

        X_bias = np.hstack((np.ones((X.shape[0], 1)), X))
        return X_bias.dot(np.hstack((self.intercept_, self.coef_)))

    def get_params(self, deep=True):
        return {"coef_": self.coef_, "intercept_": self.intercept_}

    def set_params(self, **params):
        self.coef_ = params.get("coef_", self.coef_)
        self.intercept_ = params.get("intercept_", self.intercept_)
        return self

    def mean_square_error(self, y_true, y_pred):
        return np.mean((y_true - y_pred) ** 2)

    def r_square(self, y_true, y_pred):
        ss_total = np.sum((y_true - np.mean(y_true)) ** 2)
        ss_residual = np.sum((y_true - y_pred) ** 2)
        return 1 - (ss_residual / ss_total)



# Sử dụng cross_val_score để thực hiện cross-validation
model = LinearRegressionImp()
msescores = cross_val_score(model, X_scaled, y, cv=5, scoring='neg_mean_squared_error')
rscores = cross_val_score(model, X_scaled, y, cv=5, scoring='r2')
# In kết quả các fold và giá trị trung bình
print("Average MSE: ", -np.mean(msescores))
print("R-squared scores:", np.mean(rscores))
```
